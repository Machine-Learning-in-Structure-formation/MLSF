## Converting the sets to images

```python
def dataframe_to_array(data , shape):
    ids = np.ravel_multi_index(data[['i' , 'j' , 'k']].values.T , shape)
    val = data.iloc[:,3]
    return np.bincount(ids , val , minlength = np.prod(shape)).reshape(shape)

def make_array(data, shape , column):
    mine = data.groupby(['i' , 'j' , 'k'])[column].mean().reset_index()
    arr_mine = dataframe_to_array(mine , shape)
    return mine, arr_mine
```

## Implemetnig feature and target Images:

```python
vx, arr_vx = make_array(data, (1001,1001,1001), 'vx')
print(arr_vx)
vx.head(300)
```

```python
vy, arr_vy = make_array(data, (1001,1001,1001), 'vy')
print(arr_vy)
vy.head(300)
```

```python
vz, arr_vz = make_array(data, (1001,1001,1001), 'vz')
print(arr_vz)
vz.head(300)
```

```python
In_halo, arr_In_halo = make_array(data, (1001,1001,1001), 'In_halo')
print(arr_In_halo)
In_halo.head(300)
```

```python
halo_mass, arr_halo_mass = make_array(data, (1001,1001,1001), 'halo_mass')
print(arr_halo_mass)
halo_mass.head(300)
```

## Making the X and Y matrices

```python
X = np.stack([arr_vx, arr_vy, arr_vz], axis=3)
Y = arr_In_halo

print('Features:', X)
print('Targets:', Y)
```
